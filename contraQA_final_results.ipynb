```python
import numpy as np
import pandas as pd
```

```python
rnn =[89.9, 99.9, 50.0, 52.3,
      50.7, 55.8, 50.8, 100, 49.8, 49.6, 74.9, 50.3,48.9]
gru = [98.3, 100.0, 50.2, 51.8, 52.5, 100.0, 56.8,
       100.0, 75.8, 74.3, 76.1, 80.3, 71.3]
lstm = [94.3, 100.0, 49.2, 47.5, 50.3, 100.0, 56.4,
        100.0, 74.6, 74.7, 74.2, 78.6, 70.0]
gru_inf = [98.3, 100.0, 70.7, 52.3, 87.9, 100.0, 71.1,
           100.0, 91.6, 78.3, 77.6, 83.2, 73.2]

print("rnn = {:.1f}".format(np.mean(rnn)))
print("gru = {:.1f}".format(np.mean(gru)))
print("lstm = {:.1f}".format(np.mean(lstm)))
print("gru+inf = {:.1f}".format(np.mean(gru_inf)))
```

```python
Seq2seq = [65.7, 63.2, 50.8, 52.3, 50.3, 99.8, 50.8,
           100.0, 75.5, 76.4, 76.9, 81.1, 51.4]
Seq2seqAtt = [50.9, 49.5, 50.8, 52.3, 50.3, 99.4, 50.8,
              100.0, 49.8, 74.7, 74.5, 76.8, 51.4]
MemNN = [82.1, 63.2, 50.0, 50.9, 48.7, 99.9, 55.3,
         100.0, 72.8, 72.1, 75.8, 77.2, 58.3]
MemNN_inf = [82.1, 63.2, 52.3, 54.0, 53.0, 99.9, 59.6,
            100.0, 87.9, 76.6, 75.8, 83.0, 63.9]

print("Seq2seq = {:.1f}".format(np.mean(Seq2seq)))
print("Seq2seqAtt = {:.1f}".format(np.mean(Seq2seqAtt)))
print("MemNN = {:.1f}".format(np.mean(MemNN)))
print("MemNN+inf = {:.1f}".format(np.mean(MemNN_inf)))
```

```python
gru_inf = np.array(gru_inf)
gru = np.array(gru)
rnn = np.array(rnn)
lstm = np.array(lstm)
classi = (gru_inf + gru + rnn + lstm) / 4

MemNN_inf = np.array(MemNN_inf)
MemNN = np.array(MemNN)
Seq2seq = np.array(Seq2seq)
Seq2seqAtt = np.array(Seq2seqAtt)
QA = (MemNN_inf + MemNN + Seq2seq+ Seq2seqAtt) / 4

average = (classi + QA) /2 

best_average = (MemNN_inf + gru_inf) /2 
```

```python
templates = [10, 4, 8, 2, 2, 6, 22,4, 8, 4,4,20,52]
vocab_size = [705, 706, 689, 1728, 776, 1421, 1690,734, 743, 1629,
              773, 1791, 1867]
names = ["b1", "b2", "b3", "b4", "b5", "b6", "b11", "b7" , "b8", "b9",
         "b10", "b12", "b13"]
```

```python
best_bench_class = [np.max(i) for i in zip(gru, lstm, rnn)]
best_bench_QA = [np.max(i) for i in zip(MemNN, Seq2seq, Seq2seqAtt)]
```

## Difference in classification

```python
class_diff = gru_inf - best_bench_class

class_diff = list(zip(names, class_diff))

for i in class_diff:
    print(i[0], "{:.1f}".format(i[1]))
```

## Difference in QA

```python
QA_diff = MemNN_inf - best_bench_QA

QA_diff = list(zip(names, QA_diff))

for i in QA_diff:
    print(i[0], "{:.1f}".format(i[1]))
```

```python
dic_ave ={"names": names,
      "templates": templates,
      "vocab_size": vocab_size,
      "accuracy": average}


dic_classi ={"names": names,
      "templates": templates,
      "vocab_size": vocab_size,
      "accuracy": classi}


dic_QA ={"names": names,
      "templates": templates,
      "vocab_size": vocab_size,
      "accuracy": QA}


dic_best ={"names": names,
      "templates": templates,
      "vocab_size": vocab_size,
      "accuracy": best_average}
```

```python
df_classi = pd.DataFrame(dic_classi)

df_QA = pd.DataFrame(dic_QA)

df_ave = pd.DataFrame(dic_ave)

df_best = pd.DataFrame(dic_best)


```
